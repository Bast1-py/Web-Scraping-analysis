```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
alamat = "https://indeks.kompas.com/terpopuler"
html = urlopen(alamat)
data = BeautifulSoup(html)
```

```python
import re
data.find_all(re.compile("^b"))
```

```python
for item in data:
    print(item)
```

```python
data.find_all(["a", "h2"])
```

```python
terpopuler = data.findAll("h2", {"class":"articleTitle"})
terpopuler
```

```python
Terpopuler = [result.text for result in terpopuler]
Terpopuler
```
